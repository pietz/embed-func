```python
from tokenizers import Tokenizer
import onnxruntime as ort
import numpy as np
import json
```

```python
session = ort.InferenceSession("model.onnx")
```

```python
tokenizer = Tokenizer.from_file("tokenizer.json")
```

```python
docs = ['This is an example sentence', 'This is another sentence that is slightly longer']
```

```python
embeddings = []
for doc in docs:
    tokens = tokenizer.encode(doc)
    inp = {
        "input_ids": np.array(tokens.ids, dtype=np.int64)[None, ...],
        "attention_mask": np.array(tokens.attention_mask, dtype=np.int64)[None, ...],
        "token_type_ids": np.array(tokens.type_ids, dtype=np.int64)[None, ...],
    }
    embed = session.run(None, inp)[0][0, 0]
    # embed = embed / np.linalg.norm(embed)
    embeddings.append([round(x, 4) for x in embed.tolist()])
```
